```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = 1000
flips = np.random.randint(0,2,size=n)
running_avg = np.cumsum(flips) / np.arange(1, n+1)
```

```python
plt.figure(figsize=(10, 4))
plt.plot(running_avg, color='steelblue')
plt.axhline(0.5, color='red', linestyle='--', label='True probability')
plt.xlabel('Number of flips')
plt.ylabel('Proportion of heads')
plt.title('Probability stabilizes as you flip more')
plt.legend(); plt.show()
```

```python
population = 100000
sick    = int(population * 0.01)
healthy = population - sick 
```

```python
true_positive  = int(sick * 0.99)  
false_positive = int(healthy * 0.05)
```

```python
total_positive = true_positive + false_positive
p_actually_sick = true_positive / total_positive
```

```python
print(f"Total who test positive: {total_positive}")
print(f"Actually sick among them: {true_positive}")
print(f"Real probability of being sick: {p_actually_sick:.1%}")
print(f"\nMost people guess 99%. Real answer: {p_actually_sick:.1%}")
print("Why? The disease is so rare that false alarms outnumber true cases.")
```
