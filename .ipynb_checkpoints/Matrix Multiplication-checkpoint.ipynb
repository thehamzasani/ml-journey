```python
import numpy as np 
```

```python
A = np.array([
    [1,2],
    [3,4]
])
```

```python
B = np.array([
    [5,6],
    [7,8]
])
```

```python
C = A @ B
print("A @ B = \n", C)
```

```python
print("\nManual check C[0][0]:", 1*5 + 2*7)
```

```python
stretch = np.array([[2, 0], [0, 1]])
rotation = np.array([[0, -1], [1, 0]])
```

```python
combined = rotation @ stretch
print("\nCombined transformation:\n", combined)
point = np.array([1, 0])
```

```python
print("\nOriginal point:", point)
print("After stretch then rotate:", combined @ point)
print("After rotate then stretch:", stretch @ rotation @ point)
```
