```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-3, 3, 100)
y = x ** 2
dy = 2 * x
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))
```

```python
ax1.plot(x, y, color='steelblue', lw=2)
ax1.set_title('f(x) = x²')
ax1.set_xlabel('x'); ax1.set_ylabel('f(x)')
ax1.grid(alpha=0.3)
```

```python
x0 = 2
slope = 2 * x0  # derivative at x=2
tangent = slope * (x - x0) + x0**2
ax1.plot(x, tangent, 'red', lw=1.5, label=f'tangent at x=2 (slope={slope})')
ax1.set_ylim(-1, 10)
ax1.legend()
```

```python
ax2.plot(x, dy, color='tomato', lw=2)
ax2.set_title("f'(x) = 2x  (the derivative)")
ax2.set_xlabel('x'); ax2.set_ylabel("f'(x)")
ax2.axhline(0, color='black', lw=0.5)
ax2.grid(alpha=0.3)
```

```python
plt.tight_layout(); plt.show()
```

```python

print("At x=2:  derivative =", 2*2,  "→ slope is positive → move left")
print("At x=-1: derivative =", 2*-1, "→ slope is negative → move right")
print("At x=0:  derivative =", 2*0,  "→ slope is zero     → MINIMUM!")
```
