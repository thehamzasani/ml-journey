# Matrix Transformation Visualizer
### Week 1 — Linear Algebra for ML

**The big question:** What does a matrix actually *do*?

Most people learn matrices as grids of numbers to multiply together.
But the real insight is this:

> **A matrix is a transformation. It moves space.**

In this notebook we will:
1. Build a grid of points
2. Apply different matrices to it
3. Visualize how space gets stretched, rotated, and sheared
4. Print the numbers behind each transformation

This is exactly why matrices matter for ML — neural networks chain hundreds of these transformations to reshape your data until patterns become visible.

## Step 1 — Imports

```python
import numpy as np
import matplotlib.pyplot as plt

print('Libraries loaded.')
```

## Step 2 — The helper function

This function takes a matrix and draws:
- The **original grid** in light gray
- The **transformed grid** in color — where every point moves after multiplication
- The **basis vectors** î and ĵ — the two arrows that define the entire transformation

**Key insight:** If you know where [1,0] and [0,1] land, you know everything about the matrix.

```python
def draw_grid(ax, matrix, title, color):
    """Draw original grid and transformed grid on the same axes."""

    x_vals = np.linspace(-2, 2, 9)
    y_vals = np.linspace(-2, 2, 9)

    # Original grid — faint gray background
    for x in x_vals:
        ax.axvline(x, color='lightgray', linewidth=0.5, zorder=1)
    for y in y_vals:
        ax.axhline(y, color='lightgray', linewidth=0.5, zorder=1)

    # Transformed vertical lines
    for x in x_vals:
        p1 = matrix @ np.array([x, -2])
        p2 = matrix @ np.array([x,  2])
        ax.plot([p1[0], p2[0]], [p1[1], p2[1]],
                color=color, linewidth=0.8, alpha=0.5, zorder=2)

    # Transformed horizontal lines
    for y in y_vals:
        p1 = matrix @ np.array([-2, y])
        p2 = matrix @ np.array([ 2, y])
        ax.plot([p1[0], p2[0]], [p1[1], p2[1]],
                color=color, linewidth=0.8, alpha=0.5, zorder=2)

    # Basis vectors — where does [1,0] and [0,1] land?
    origin = np.array([0, 0])
    i_hat  = matrix @ np.array([1, 0])
    j_hat  = matrix @ np.array([0, 1])

    ax.annotate('', xy=i_hat, xytext=origin,
                arrowprops=dict(arrowstyle='->', color='#E24B4A', lw=2.5))
    ax.annotate('', xy=j_hat, xytext=origin,
                arrowprops=dict(arrowstyle='->', color='#378ADD', lw=2.5))

    ax.text(i_hat[0]+0.1, i_hat[1]+0.1, 'î', color='#E24B4A',
            fontsize=13, fontweight='bold')
    ax.text(j_hat[0]+0.1, j_hat[1]+0.1, 'ĵ', color='#378ADD',
            fontsize=13, fontweight='bold')

    # Axes
    ax.axhline(0, color='#333', linewidth=0.8, zorder=3)
    ax.axvline(0, color='#333', linewidth=0.8, zorder=3)

    ax.set_xlim(-3, 3)
    ax.set_ylim(-3, 3)
    ax.set_aspect('equal')
    ax.set_title(title, fontsize=13, fontweight='bold', pad=10)
    ax.grid(False)

print('Function defined.')
```

## Step 3 — Define the 3 matrices

Each matrix transforms space in a completely different way.

| Matrix | What it does |
|--------|-------------|
| Stretch | Pulls space wider horizontally |
| Rotation | Spins the entire grid |
| Shear | Pushes the top sideways — like leaning a rectangle |

```python
# 1. Stretch — doubles x, leaves y alone
stretch = np.array([
    [2, 0],
    [0, 1]
])

# 2. Rotation — rotates 45 degrees anticlockwise
angle = np.radians(45)
rotation = np.array([
    [np.cos(angle), -np.sin(angle)],
    [np.sin(angle),  np.cos(angle)]
])

# 3. Shear — slides the top of space sideways
shear = np.array([
    [1, 1],
    [0, 1]
])

print('Stretch matrix:')
print(stretch)
print('\nRotation matrix (45 deg):')
print(np.round(rotation, 2))
print('\nShear matrix:')
print(shear)
```

## Step 4 — Visualize all 3 transformations

Run the cell below. You will see three panels:
- The gray background is where the grid **was**
- The colored lines show where the grid **moved to**
- The red and blue arrows show the basis vectors

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 5))
fig.suptitle('What does a matrix actually do to space?',
             fontsize=15, fontweight='bold', y=1.02)

configs = [
    (stretch,  'Stretch matrix\n[[2,0],[0,1]]',  '#1D9E75'),
    (rotation, 'Rotation matrix\n(45 degrees)',   '#7F77DD'),
    (shear,    'Shear matrix\n[[1,1],[0,1]]',     '#D85A30'),
]

for ax, (matrix, title, color) in zip(axes, configs):
    draw_grid(ax, matrix, title, color)

plt.tight_layout()
plt.savefig('matrix_transformation.png', dpi=150, bbox_inches='tight')
plt.show()
print('Plot saved as matrix_transformation.png')
```

## Step 5 — See the numbers behind the visuals

Every transformation above is just matrix multiplication.
Here you can see exactly what happens to specific points.

```python
points = [[1, 0], [0, 1], [1, 1], [2, 1]]

print('--- Stretch matrix: what happens to each point ---')
for p in points:
    result = stretch @ np.array(p)
    print(f'  {p}  →  {result.tolist()}')

print('\n--- Rotation matrix: what happens to each point ---')
for p in points:
    result = rotation @ np.array(p)
    print(f'  {p}  →  [{result[0]:.2f}, {result[1]:.2f}]')

print('\n--- Shear matrix: what happens to each point ---')
for p in points:
    result = shear @ np.array(p)
    print(f'  {p}  →  {result.tolist()}')
```

## Step 6 — Experiment yourself

Try changing the matrix values below and rerun to see what happens.
This is the fastest way to build intuition.

```python
# Change these numbers and rerun — what happens to space?
my_matrix = np.array([
    [1, 2],   # try: [2,0], [0,2] — what does that do?
    [0, 1]    # try: [-1,0], [0,1] — what does that do?
])

fig, ax = plt.subplots(figsize=(5, 5))
draw_grid(ax, my_matrix, f'My matrix\n{my_matrix.tolist()}', '#BA7517')
plt.tight_layout()
plt.show()

print('Matrix used:')
print(my_matrix)
print('\nWhere does [1,0] go?', my_matrix @ np.array([1, 0]))
print('Where does [0,1] go?', my_matrix @ np.array([0, 1]))
```

## Key takeaways

1. **A matrix is a transformation** — it moves every point in space to a new location
2. **Basis vectors tell the whole story** — where [1,0] and [0,1] land completely defines the matrix
3. **Neural networks chain these transformations** — layer by layer, they reshape your data until classes become linearly separable
4. **Matrix multiplication = applying transformations in sequence** — this is what happens inside every forward pass
