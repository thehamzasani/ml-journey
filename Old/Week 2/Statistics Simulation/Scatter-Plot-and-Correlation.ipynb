```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
hours_studied = np.random.uniform(1, 10, 100)
exam_score = hours_studied * 8 + np.random.normal(0, 5, 100)
```

```python
plt.figure(figsize=(8,5))
plt.scatter(hours_studied, exam_score, alpha=0.6, color='steelblue')
plt.xlabel('Hours studied')
plt.ylabel('Exam score')
plt.title('Correlation between study time and score')
plt.show()
```

```python
corr = np.corrcoef(hours_studied, exam_score)[0,1]
print(f"Correlation: {corr:.3f}")
```
