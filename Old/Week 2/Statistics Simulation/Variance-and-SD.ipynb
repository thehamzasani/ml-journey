```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
scores = np.random.normal(loc=70, scale=10, size=1000)
# loc = mean (average), scale = std deviation (spread)

print(f"Mean:   {np.mean(scores):.2f}")
print(f"Std:    {np.std(scores):.2f}")
print(f"Min:    {np.min(scores):.2f}")
print(f"Max:    {np.max(scores):.2f}")

# Plot the distribution
plt.hist(scores, bins=40, color='steelblue', edgecolor='white')
plt.axvline(np.mean(scores), color='red', label='Mean')
plt.title('Normal distribution of exam scores')
plt.legend()
plt.show()
# Change loc and scale — watch the shape shif
```
