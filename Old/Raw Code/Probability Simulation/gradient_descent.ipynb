```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    return x ** 2 

def df(x):
    return 2 * x  

x = 3.0 
lr = 0.1 
steps = 20  

history_x = [x]
history_y = [f(x)]
```

```python
for i in range(steps):
    slope = df(x)            # what is the slope right here?
    x = x - lr * slope       # move a little bit opposite to slope
    history_x.append(x)
    history_y.append(f(x))
    print(f"Step {i+1:2d} | x = {x:.5f} | f(x) = {f(x):.5f} | slope = {slope:.5f}")
```

```python
x_curve = np.linspace(-4, 4, 200)
y_curve = f(x_curve)

plt.figure(figsize=(8, 5))
plt.plot(x_curve, y_curve, color='steelblue', linewidth=2, label='f(x) = x²')
plt.plot(history_x, history_y, 'o--', color='orange',
         linewidth=1.5, markersize=6, label='gradient descent steps')
plt.scatter(history_x[0], history_y[0], color='red', s=100, zorder=5, label='start')
plt.scatter(history_x[-1], history_y[-1], color='green', s=100, zorder=5, label='end')
plt.title('Gradient Descent from Scratch')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```

```python
point = 3.0        # change this to 1.0, 0.5, 0.0 and rerun

slope = df(point)  # slope at that point

# tangent line — a straight line that touches the curve at your point
# y = slope * (x - point) + f(point)  is the formula for a tangent line
x_line = np.linspace(point - 1.5, point + 1.5, 100)
y_line = slope * (x_line - point) + f(point)

# the curve
x_curve = np.linspace(-4, 4, 200)
y_curve = f(x_curve)

plt.figure(figsize=(8, 5))
plt.plot(x_curve, y_curve, color='steelblue', linewidth=2.5, label='f(x) = x²')
plt.plot(x_line, y_line, color='red', linewidth=2,
         linestyle='--', label=f'slope at x={point} is {slope}')
plt.scatter([point], [f(point)], color='orange',
            s=120, zorder=5, label='your point')

plt.title(f'Slope at x = {point}')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```
