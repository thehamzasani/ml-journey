```python
n = 10000
flips = np.random.randint(0,2, size=n)
running_avg = np.cumsum(flips) / np.arange(1, n+1)
```

```python
plt.figure(figsize=(10,2))
plt.plot(running_avg, color='steelblue')
plt.axhline(y=0.5, color='red', linestyle='--', label='True probability = 0.5')
plt.xlabel('Number of flips')
plt.ylabel('Proportion of heads')
plt.title('Probability stabilizes as sample size grows')
plt.legend()
plt.show()
```

```python
numOfTrials = 10000
coinFlips = np.random.randint(0, 2, size=numOfTrials)
# Probability of heads
numberOfHeads = np.sum(coinFlips == 1)
print(f"Total number of heads in {numOfTrials} flips: {numberOfHeads}")
probabilityOfHeads = numberOfHeads / numOfTrials
print(f"Estimated probability of heads: {probabilityOfHeads:.3f}")
# Probability of tails
numOfTails = np.sum(coinFlips == 0)
print(f"Total number of tails in {numOfTrials} flips: {numOfTails}")
probabilityOfTails = numOfTails / numOfTrials
print(f"Estimated probability of tails: {probabilityOfTails:.3f}")
```

```python
firstCoin = np.random.randint(0, 2, size=numOfTrials)
secondCoin = np.random.randint(0, 2, size=numOfTrials)

probabilityOfSecondIsHead = np.sum(secondCoin == 1) / numOfTrials
print(f"Estimated probability of second coin being heads: {probabilityOfSecondIsHead:.3f}")

probabilityOfFirstIsHead = np.sum(firstCoin == 1) / numOfTrials
print(f"Estimated probability of first coin being heads: {probabilityOfFirstIsHead:.3f}")
```

```python
# probability of second and first is h
bothHeads = np.sum((firstCoin == 1) & (secondCoin == 1)) / numOfTrials
conditionalProabilityOfBGivenA = bothHeads / probabilityOfFirstIsHead
print(f"Estimated probability of second coin being heads given first coin is heads: {conditionalProabilityOfBGivenA:.3f}")

probabilityOfSecondIsHead = np.sum(secondCoin == 1) / numOfTrials
print(f"Estimated probability of second coin being heads: {probabilityOfSecondIsHead:.3f}")
```

```python
p_A_g_B = (np.sum((firstCoin == 1) & (secondCoin == 1)) / numOfTrials) / (np.sum(secondCoin == 1) / numOfTrials)
print(f"Estimated probability of first coin being heads given second coin is heads: {p_A_g_B:.3f}")


bothHeads = np.sum((firstCoin == 1) & (secondCoin == 1)) / numOfTrials
print(f"Estimated probability of both coins being heads: {bothHeads:.3f}")
```

```python
def bayes_disease(population=100000):
    p_disease     = 0.01   # 1% have disease
    p_pos_given_sick    = 0.99   # sensitivity
    p_pos_given_healthy = 0.05   # false positive rate

    sick    = int(population * p_disease)
    print(f"Number of sick people: {sick}")
    healthy = population - sick
    print(f"Number of healthy people: {healthy}")

    true_positives  = int(sick * p_pos_given_sick)
    false_positives = int(healthy * p_pos_given_healthy)

    total_positives = true_positives + false_positives
    p_sick_given_pos = true_positives / total_positives

    print(f"Population:       {population}")
    print(f"Actually sick:    {sick}")
    print(f"Test positives:   {total_positives}")
    print(f"True positives:   {true_positives}")
    print(f"False positives:  {false_positives}")
    print(f"\nP(actually sick | test positive) = {p_sick_given_pos:.2%}")
    print("Most people are shocked by this number.")

bayes_disease()
```

```python
# Function
def f(x):
    return x**2

# Derivative
def df(x):
    return 2*x

# Gradient Descent
x = 8          # starting point (far from minimum)
lr = 0.1       # learning rate
steps = 20

x_history = [x]

for i in range(steps):
    x = x - lr * df(x)
    x_history.append(x)

# Plot curve
x_vals = np.linspace(-10, 10, 100)
y_vals = f(x_vals)

plt.plot(x_vals, y_vals)

# Plot steps
y_history = [f(x) for x in x_history]
plt.scatter(x_history, y_history)

plt.title("Gradient Descent on f(x) = x^2")
plt.show()
```
