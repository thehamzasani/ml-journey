```python
import math
from sympy import *
```

```python
P0 = 10      
r = 0.4      
```

```python
for t in range(0, 11):
    P = P0 * math.exp(r * t)
    print(f"t={t} hours -> population={P:.2f}")
```

```python
x = symbols('x')
f = 1/x
result = limit(f, x , oo)
print(result)
```

```python
n = symbols('n')
f = (1 + (1/n))**n
result = limit(f, n, oo)
```

```python
print(result) 
print(result.evalf()) 
```
