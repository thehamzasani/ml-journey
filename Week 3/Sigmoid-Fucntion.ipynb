```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
def sigmoid(z):
    return 1 / (1+np.exp(-z))
```

```python
print(sigmoid(100))
```

```python
print(sigmoid(0))
```

```python
print(sigmoid(10))   
print(sigmoid(-10)) 
```

```python
z = np.linspace(-10, 10, 300)
plt.plot(z, sigmoid(z), color='steelblue', linewidth=2.5)
plt.axhline(0.5, color='red', linestyle='--', label='Decision boundary')
plt.title('Sigmoid: any number → probability')
plt.legend(); plt.show()
```
