```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)
```

```python
hours = np.random.uniform(1, 10, 100)
passed = (hours * 8 + np.random.normal(0, 5, 100) > 50).astype(int)
```

```python
x = hours.reshape(-1, 1)
y = passed
```

```python
def sigmoid(z):
    return 1 / (1+np.exp(-z))
```

```python
w = np.zeros(1)
b = 0.0
lr = 0.1
n = len(x)
losses = []
```

```python
for epoch in range(300):
    z = x.dot(w) + b 
    y_pred = sigmoid(z)
    error = y_pred - y 

    y_c = np.clip(y_pred, 1e-7, 1-1e-7)
    loss = -np.mean(y*np.log(y_c) + (1-y)*np.log(1-y_c))
    losses.append(loss)

    dw = (1/n) * x.T.dot(error)
    db = (1/n) * np.sum(error)
    w -= lr * dw.flatten()
    b -= lr * db 

    if epoch % 60 == 0:
        acc = np.mean((y_pred > 0.5).astype(int) == y)
        print(f"Epoch {epoch:3d} | Loss: {loss:.4f} | Accuracy: {acc:.2%}")
```
