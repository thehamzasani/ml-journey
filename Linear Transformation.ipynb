```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
stretch = np.array([
    [2,0],
    [0,1]
])
```

```python
angle = np.radians(45)
rotation = np.array([
    [np.cos(angle), -np.sin(angle)],
    [np.sin(angle), np.cos(angle)]
])
```

```python
shear = np.array([
    [1,1],
    [0,1]
])
```

```python
def show_transform(matrix, title):
    fig, ax = plt.subplots(figsize=(5, 5))
    
    # Draw transformed grid lines
    for x in np.linspace(-2, 2, 9):
        p1 = matrix @ np.array([x, -2])
        p2 = matrix @ np.array([x,  2])
        ax.plot([p1[0], p2[0]], [p1[1], p2[1]], 
                'steelblue', alpha=0.5)
    
    for y in np.linspace(-2, 2, 9):
        p1 = matrix @ np.array([-2, y])
        p2 = matrix @ np.array([ 2, y])
        ax.plot([p1[0], p2[0]], [p1[1], p2[1]], 
                'steelblue', alpha=0.5)
    
    ax.set_xlim(-3, 3); ax.set_ylim(-3, 3)
    ax.axhline(0, color='black', lw=0.5)
    ax.axvline(0, color='black', lw=0.5)
    ax.set_aspect('equal')
    ax.set_title(title)
    plt.show()
```

```python
show_transform(stretch, 'Stretch')
show_transform(rotation, 'Rotation 45°')
show_transform(shear, 'Shear')
```
